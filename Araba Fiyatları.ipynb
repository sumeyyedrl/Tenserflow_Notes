### model oluşturmak

```python
y=dataFrame["price"].values
x=dataFrame.drop("price",axis=1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model= Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit( x=x_train, y=y_train, validation_data=(x_test,y_test), batch_size=250, epochs=300)
```

```python
kayipVeri=pd.DataFrame(model.history.history)
```

```python
kayipVeri.head()
```

```python
kayipVeri.plot()
```

### test datasını işleme sokmak

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
tahminDizisi=model.predict(x_test) #testteki verilere göre fiyatlar tahmin edilir ve diziye atanır
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi) #ortalama sapma
```
