```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yasListesi=[10,20,30,30,30,40,50,60,70,75]
kiloListesi=[20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYas=np.array(yasListesi)
numpyKilo=np.array(kiloListesi)
```

```python
plt.plot(numpyYas,numpyKilo,"g") 
# x=yas y=kilo g=green
```

```python
plt.plot(numpyKilo,numpyYas,"b") 
plt.xlabel("Kilo")
plt.ylabel("Yas")
plt.title("Yas-Kilo Grafiği")
```

```python
dizi1=np.linspace(0,10,20)
dizi2=dizi1**2
```

```python
plt.subplot(1,2,1) # 1 sıra, 2 kolon, 1. grafik
plt.plot(dizi1,dizi2,"r*-")
plt.subplot(1,2,2) # 1 sıra, 2 kolon, 2. grafik
plt.plot(dizi2,dizi1,"y--")
```

```python
figure1=plt.figure() #obje oluşturulur. fonksiyon sadece figür döndürür
eksen=figure1.add_axes([0.1,0.1,0.6,0.6]) #0.6 = width,height
eksen.plot(dizi1,dizi2,"black")
eksen.set_xlabel("x ekseni")
eksen.set_ylabel("y ekseni")
eksen.set_title("başlık")
```

```python
figure2=plt.figure()
eksen1= figure2.add_axes([0.1,0.1,0.9,0.9]) 
eksen2= figure2.add_axes([0.2,0.6,0.3,0.3]) #0.2,0.6=koordinatlar

eksen1.plot(dizi1,dizi2,"pink")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("büyük grafik")

eksen2.plot(dizi2,dizi1,"brown")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("küçük grafik")
```

```python
benimFigure,benimEksen=plt.subplots() #fonksiyon hem figür hem de eksen döndrür
```

```python
benimFigure,benimEksen=plt.subplots(nrows=1,ncols=2)

for eksen in benimEksen: #benimEksen ndarray döndürdüğünden for loop gerekir
    eksen.plot(dizi1,dizi2,"orange")
    eksen.set_xlabel("X ekseni")
    
plt.tight_layout() #grafiklerin birbirine girmesini engeller
```

```python
figure3=plt.figure()
# figure3=plt.figure(figsize=(6,6)) width,height değeri, add_axes'i overwright eder
# figure3=plt.figure(dpi=100) grafiğin kalitesini belirler
eksen3=figure3.add_axes([0.1,0.1,0.9,0.9])
eksen3.plot(dizi1,dizi1**2,label="dizi1**2",color="#f36cac")
eksen3.plot(dizi1,dizi1**3,label="dizi1**3",color="#f36cac",alpha=0.3)
eksen3.legend(loc=6) # lokasyon değiştirilebilir

# figure3.savefig("benimGrafiğim",dpi200)  
```

```python
figure4=plt.figure()
eksen4=figure4.add_axes([0.1,0.1,2.0,2.0])
eksen4.plot(dizi1,dizi1+2,label="linewidth=1.0",color="blue",linewidth=1.0)
eksen4.plot(dizi1,dizi1+3,label="linewidth=3.0",color="red",linewidth=3.0)

eksen4.plot(dizi1,dizi1+4,label="linestyle= --",color="green",linestyle="--")
eksen4.plot(dizi1,dizi1+5,label="linestyle= :",color="grey",linestyle=":")
eksen4.plot(dizi1,dizi1+6,label="linestyle= -.",color="pink",linestyle="-.")

eksen4.plot(dizi1,dizi1+7,label="marker= +",color="#e845c0",marker="+")
eksen4.plot(dizi1,dizi1+8,label="marker= *",color="#f36cac",marker="*")
eksen4.plot(dizi1,dizi1+9,label="marker= o",color="#e8a258",marker="o")

eksen4.plot(dizi1,dizi1+10,label="markersize=8",color="purple",marker="o",markersize=8)
eksen4.plot(dizi1,dizi1+11,label="markerfacecolor= red",color="black",marker="o",markerfacecolor="red")


eksen4.legend()
```
